#### **Exercise 1:**
Create a function returning the first N terms of the tailor series of the function without using the sympy series function.

```python
# Libraries
import numpy as np
import sympy as sp
import math
from matplotlib import pyplot as plt
from scipy import optimize
from scipy.integrate import quad
```

```python
def taylor_diff(x, expr, point, n):
    t = []
    for i in range(n):
        if i > 0:
            expr = sp.diff(expr, x)
        diff_point = expr.subs(x, point)
        term = (diff_point / sp.factorial(i))*(x - point)**i
        t.append(term)
    return sp.simplify(sum(t))
```

```python
x = sp.symbols('x')
sol = taylor_diff(x, sp.sin(x), 0, 5)
sol
```

```python
sp.sin(x).series(x, point, n)
```

```python
sol = taylor_diff(x, sp.cos(x), 0, 5)
sol
```

```python
sp.cos(x).series(x, point, n)
```

#### **Exercise 2:**

Write a matrix multiplication function using lists ("manually", using the matrix multiplication formula) and another one using numpy arrays (call matmul function). Time them both. 

```python
def matrix_multiplication(a, b):
    m = len(a)
    n = len(a[0])
    p = len(b[0])
    if n == len(b):
        c = np.zeros((m, p), dtype='int')
        for i in range(m):
            for j in range(p):
                for k in range(n):
                    c[i][j] += a[i][k]*b[k][j]
    else:
        print("The multiplication is not possible")
    return c
```

```python
a = [[1, 2, 4],[2, 4, 6],[6, 7 ,8]]
b = [[3, 4, 9],[3, 2, 8],[7, 8, 1]]

c1 = matrix_multiplication(a, b)
print(c1)
```

```python
c2 = np.matmul(a, b)
print(c2)
```

```python
%timeit c1 = matrix_multiplication(a, b)
```

```python
%timeit c2 = np.matmul(a, b)
```

#### **Exercise 4:**
Plot the recaman sequence in the following way:

Add a semi-circular arc from $a_{n}$ to $a_{n-1}$.

Alternate the arcs to be above and below the x-axis according to whether
    n is even or odd.

```python
def recaman(n):
    a = []
    a.append(0)
    s = set()
    s.add(0)
    for i in range(1, n):
        x = a[i-1] - i
        if (x > 0 and x not in s):
            a.append(x)
            s.add(x)
        else:
            x = a[i-1] + i
            a.append(x)
            s.add(x)
    return a
```

```python
a = recaman(30)

fig = plt.figure(figsize=(8, 6))
for i in range(1, len(a)):
    x1 = a[i-1]
    x2 = a[i]
    center = (x1 + x2)/2
    radius = abs(x2 - x1)/2

    t = np.linspace(0, np.pi, 100)
    x = center + radius*np.cos(t)

    sign = 1 if (i%2 == 0) else -1
    y = sign*radius*np.sin(t)

    plt.plot(x, y, color='b', lw=1)

plt.axhline(0, color='black')
plt.title('Recamán Function',fontsize = 'large', fontweight='bold')
plt.ylabel('y',fontstyle='italic')
plt.xlabel('x',fontstyle='italic')
plt.tight_layout()
plt.show()
```

#### **Exercise 5:**

Plot a rainbow. Use your imagination and google to make it beautiful.

```python
theta = np.linspace(0, np.pi, 600)

r_out = 1.0
r_in = 0.5

colors = plt.cm.rainbow(np.linspace(0, 1, len(theta)))

plt.figure(figsize=(10, 6))

for i in range(len(theta)-1):
    t = [theta[i], theta[i+1]]

    x_out = r_out*np.cos(t)
    y_out = r_out*np.sin(t)

    x_in = r_in*np.cos(t[::-1])
    y_in = r_in*np.sin(t[::-1])

    plt.fill(np.concatenate([x_out, x_in]), np.concatenate([y_out, y_in]), color = colors[i], lw=0)

plt.title("Rainbow")
plt.show()
```

#### **Exercise 6:**

Plot a nice-looking 3-D torus. Do this 3 times with different styles. 

```python
R = 3.0
r = 1.0

u = np.linspace(0, 2*np.pi, 120)
v = np.linspace(0, 2*np.pi, 60)
U, V = np.meshgrid(u, v)

X = (R + r*np.cos(V)) * np.cos(U)
Y = (R + r*np.cos(V)) * np.sin(U)
Z = r*np.sin(V)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, linewidth=0, antialiased=True)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

ax.set_xlim(-r-R, r+R)
ax.set_ylim(-r-R, r+R)
ax.set_zlim(-r-R, r+R)
plt.show()
```

#### **Exercise 7:**

Repeat the curve fitting example (from SciPy lecture) with the following modifications:

 - choose some other function
 - first write the data to a file, then read it (with numpy, for example)
 - produce the picture that has axes, name and legend


```python
np.random.seed(0)

# Let's generate fake data that we can "fit" then 
def f(x, a, b, c):
    return a*np.exp(-b*x)*np.cos(c*x)

#random makes it look like experimental data
x = np.linspace(0, 10, 80)
a = 2.5
b = 0.25
c = 2.0
y = f(x, a, b, c)+ .4*np.random.normal(size=80)

# Save the data
data1 = np.column_stack([x, y])
np.savetxt("./fit_data.txt", data1)

# Read the data
data2 = np.loadtxt("./fit_data.txt")
x2 = data2[:, 0]
y2 = data2[:, 1]

# Now fit the model: the parameters omega and phi can be found in the
# `params` vector
params, params_cov = optimize.curve_fit(f, x2, y2)

print(params)

# plot the data and the fitted curve
plt.plot(x2, y2, 'bx')
plt.plot(x2,f(x2, a, b, c),'g-')
plt.plot(x2, f(x2, *params), 'r-')
```

```python
def f(x):
    return x**2 + 10*np.sin(x)

result, error = quad(f, -10, 10)
print("Result:", result)
print("Error:", error)
```
